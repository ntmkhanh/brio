```python

key = "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"

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("ynhi/vit5-finetune-v6", use_auth_token=key)

model = AutoModelForSeq2SeqLM.from_pretrained("ynhi/vit5-finetune-v6", use_auth_token=key)
```

```python
import nltk
nltk.download('punkt')
```

```python
text = """
thịt heo, tỏi, tiêu, đường, bột ngọt, nước mắm, hạt nêm
"""

inputs = "công thức: " + text

inputs = tokenizer(inputs, max_length=128, truncation=True, return_tensors="pt")
output = model.generate(**inputs, num_beams=5, do_sample=True, min_length=22, max_length=1024, no_repeat_ngram_size=3,)
decoded_output = tokenizer.batch_decode(output, skip_special_tokens=False)[0]
predicted_recipe = nltk.sent_tokenize(decoded_output.strip())
predicted_string = ' '.join(predicted_recipe)

print(predicted_string)
```
